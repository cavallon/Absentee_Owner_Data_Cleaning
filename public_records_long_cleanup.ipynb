```python
public_data = pd.read_csv(file_path)
public_data.head()
```

```python
df = public_data[['Owner First Name', 'Owner Last Name', 'Property Address Formatted', 'City', 'County', 'Postal Code', 'Owner Address', 'Owner City State', 'Owner Postal Code']]
```

```python
df['Property State'] = df['County'].str[-2:]
```

```python
df = df.drop(columns=['County'])
```

```python
df['Postal Code'] = df['Postal Code'].astype(str).str.zfill(5)
```

```python
df['Mailing State'] = df['Owner City State'].str[-2:]  # Get the last two characters for the state
df['Mailing City'] = df['Owner City State'].str[:-3]
```

```python
df = df.drop(columns=['Owner City State'])
```

```python
df['Owner Postal Code'] = df['Owner Postal Code'].astype(str).str.replace('.0', '', regex=False)
```

```python
df['Owner Postal Code'] = df['Owner Postal Code'].astype(str).str.zfill(5)
```

```python
column_counts = df.count()
print(column_counts)
```

```python
df = df[['Owner First Name', 'Owner Last Name', 'Property Address Formatted', 'City', 'Property State', 'Postal Code', 'Owner Address', 'Mailing City', 'Mailing State', 'Owner Postal Code']]
```

```python
df = df.rename(columns={
    'Property Address Formatted': 'Property Address',
    'City': 'Property City',
    'Postal Code': 'Property Zip Code', 
    'Owner Address': 'Mailing Address',
    'Owner Postal Code': 'Mailing Zip Code'

})
```

```python
df
```

```python
# Combine first and last names to check for duplicates
df['full_name'] = df['Owner First Name'] + ' ' + df['Owner Last Name']
```

```python
# Combine Mailing Address and City to check for duplicates
df['Address_City'] = df['Mailing Address'] + ' ' + df['Mailing City'].str.lower().str.strip()
```

```python
# Identify duplicates based on a subset of columns
duplicates = df[df.duplicated(subset=['Address_City'], keep='first')]

# View duplicates
print(duplicates)
```

```python
df.drop_duplicates(subset='Address_City', keep='first', inplace=True)
```

```python
#Drop all duplicates except for the first occurance:

df.drop_duplicates(subset='full_name', keep='first', inplace=True)
```

```python
# Drop the combined column if no longer needed
df = df.drop(columns=['Address_City'])
```

```python
# Drop the combined column if no longer needed
df = df.drop(columns=['full_name'])
```

```python
df.to_csv('Lawrence_Society_Hill_Owners-with-duplicates.csv', index=False)
```
